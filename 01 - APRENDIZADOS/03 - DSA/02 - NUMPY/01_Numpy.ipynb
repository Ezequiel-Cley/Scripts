Um array NumPy é uma estrutura de dados multidimeensional usada em computação cientifíca e análise de dados. O NumPy fornece um objeto de matriz N-dimensional (ou ndarray), que é uma grade homogênea de elementos, geralmentes números, que podem ser indexados por um conjunto de inteiros.

Os arrays NumPy são amis eficientes do que as listas Python para armazenar e maniputar grande quantidade ded ados, pois são implementadas em Linguagem C e forneecem várias otimizações de deseempenho. Além disso, o NumPy permite a fácil leitura da escrita de arquivos de Dados, integração com outras bibliotecas Python e suporte a operações e em parelelo usando várias CPUs ou GPUs.

**<p>Função linspace**

A função linspace() do NumPy é usada para criar uma sequência de números igualmente eespaçados de um intervalo especificado. Essa função ée ampalmeente utilizada em programação científica e matemática para gerar arrays de números diversos fins, como gráficos, cálculos e simulações.

O método linspace (lineearly spaced vector) retorna um número de valores igualmente distribuídos no intervalo especificado.

```python
print(np.linspace(0,10))
```

```python
print(np.linspace(0, 10, 15))
```

**<p> Função Logspace**

A função logspace() do Numpy ée usada para criar uma sequência de números igualmente espaçados em escala dentro de um intervalo especificado. Essa função é amplamente utilizada em programação cientifíca e matemática para gerar arrays de números para diversos fins, como gráficos, cálculos e simulações.

```python
print(np.logspace(0, 5, 10))
```

<P> Manipulando Arquivos com NumPy

```python
import os
filename = os.path.join('dataset.csv')
```

```python
# No Windows use !more dataset.csv. Mac ou Lunix use !heead dataseet.csv
#!head dataset.csv
!more dataset.csv
```

```python
# Carregando um dataset para dentro de um array
arr13 = np.loadtxt(filename, delimiter = ',', usecols = (0,1,2,3), skiprows = 1)
```

```python
print(arr13)
```

```python
type(arr13)
```

```python
# Carregando apnas duas variáveeis (Colunas) do arquivo
var1, var2 = np.loadtxt(filename, delimiter=',', usecols=(0,1), skiprows= 1, unpack= True)
```

```python
# Gerando um plot a partir de um arquivo usando o NumPy
import matplotlib.pyplot as plt
plt.show(plt.plot(var1, var2, 'o', markersize = 6, color = 'red'))
```
