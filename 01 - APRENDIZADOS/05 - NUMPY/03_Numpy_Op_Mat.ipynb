<p> A função dot()

Para multiplicar duas matrizes NumPymm podemos usar a função dot() ou o operador @. Ambos os métodos executam a multiplicação matricial. É importante lemmbrar que, para a multiplicação de matrizes possa ser executada, o número dee colunas da primeira matriz deve ser igual ao número de linhas da segunda matriz.

Há várias formas dee multiplicar elementos de matrizees NumPy. A função dot() é um método bastante utilizado.

```python
# Cria duas matrizes
arr19 = np.array([[1,2],[3,4]])
arr20 = np.array([[5,6],[0,7]])
```

```python
print(arr19.shape,arr20.shape)
```

```python
print(arr19)
```

```python
print(arr20)
```

```python
# multiplicar as duas Matrizes 
arr21 = np.dot(arr19, arr20)
print(arr21)
```

```python
# Multiplicar as duas matrizes
arr21 = arr19 @ arr20
print(arr21)
```

```python
# Multiplicar as duas matrizes
arr21 = np.tensordot(arr19,arr20, axes = ((1),(0)))
print(arr21)
```

<p> Método flatten()

O método flatten() com NumPy é usado para criar uma cópia unidimensional (ou "achatada") de um array multidimensional. Isso significa que o método cria um novo array undimensional, que contém todos os elementos do array multidimensional original, mas que está organizado em uma única linha. A ordem dos elementos no novo array undimensional segue a ordem dos eelementos no array multidimensional original.

```python
# Achatando a matriz
arr27 = arr26.flatten()
print(arr27)
```
