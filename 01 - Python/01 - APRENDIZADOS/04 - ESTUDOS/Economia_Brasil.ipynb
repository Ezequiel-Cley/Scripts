```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo a base de Dados
Base_Dados = pd.read_excel('Dados_Pib.xlsx')

#Verificando
Base_Dados.head()
```

```python
Base_Dados.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
# Sistema de Grids

# Cor Fundo e Texto
cor_fundo = '#f2f5fa'
cor_texto = '#000000'

# Criando o sistema de Grids
Grid_Graficos = sns.FacetGrid(Base_Dados,col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Aficionar grafico linhas em cada Grid
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

# Adicionar Sombras + Ajustes de um Titulo
Grid_Graficos = Grid_Graficos.map( plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name}')

# Adicionar um Subtitulo
Grid_Graficos = Grid_Graficos.fig.suptitle(
    "Evolução de Renda Per Capita \n Esse relatório foi elaborado para apenas treinamento",
    fontsize=18, color=cor_texto
    )

# Ajustando o espaçamento do titulo
plt.subplots_adjust( top= 0.92)
```
