```python
# Importando bibliotecas
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
```

```python
# lendo a base de dados
df = pd.read_excel(r'C:\Users\ezequ\Scripts\01 - Python\01 - APRENDIZADOS\04 - ESTUDOS\Dados_Bitcoin.xlsx')

# Verificando Base de Dados
df.head()
```

```python
# Grafico de linhas
fig = px.line(df,  y='Close', title='Bitcoin')
# Setar o index
df.set_index('Date', inplace=True)
# Mostrar 
fig.show()
```

```python
# Gerar medias moveis
df['Media_Movel'] = df['Close'].rolling(5).mean()
df['Media_Tedencia'] = df['Close'].rolling(30).mean()
```

```python
# Criando Dashboard

# Criando uma figura
fig = go.Figure()

# Adicionar grficos
fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df.Close,
        mode = 'lines',
        name = 'Fechamento',
        marker_color = 'Orange',
        opacity=0.5,
    )
)

# Adicionando a mdia
fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df.Media_Movel,
        mode = 'lines',
        name = 'Mdia Mvel',
        marker_color = 'Blue',
        opacity=0.5,
    )
)

# Adicionando a mdia por tedencia
fig.add_trace(
    go.Scatter(
        x = df.index,
        y = df.Media_Tedencia,
        mode = 'lines',
        name = 'Tedncia',
        marker_color = 'Red',
        opacity=0.5,
    )
)

# Ajustar Layout
fig.update_layout(
    
    # Titulo
    title = "Anlise do fechamento Bitcoin",
    # Tamanho
    #titlefont_size=20,
    
    xaxis = dict(
        title='Periodo Histrico',
        tickfont_size=8
        #titlefont_size = 14
    ),
    yaxis = dict(
        title='Preo Fechamento ($)',
        #titlefont_size = 14
    ),
    
    # Parametros para legenda
    legend = dict(
        y=1,
        x=1,
        
    )
)
```
